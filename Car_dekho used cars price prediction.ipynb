```python
df = df[['vehicle_age','km_driven','mileage','engine','max_power','seats','seller_type_Dealer','seller_type_Individual','seller_type_Trustmark Dealer','fuel_type_CNG','fuel_type_Diesel','fuel_type_Electric','fuel_type_LPG','fuel_type_Petrol','transmission_type_Automatic','transmission_type_Manual','selling_price']]
df
```

```python
x=df.drop('selling_price',axis=1)  
x
```

```python
y=df['selling_price']   
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=225)
```

```python
plt.figure(figsize=(14,8))
sns.heatmap(x_test.corr(),annot=True)
plt.show()
```

```python
x_test=x_test.drop(['max_power', 'seats', 'fuel_type_Diesel'],axis=1)
x_test
```

```python
plt.figure(figsize=(14,8))
sns.heatmap(x_test.corr(),annot=True,cmap='hot_r')
plt.show()
```

```python
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import ExtraTreesRegressor, RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, explained_variance_score, r2_score
```

```python
models = [ LinearRegression, SVR, DecisionTreeRegressor, RandomForestRegressor, Ridge, Lasso]
mse = []
rmse = []
ex_var_score = []
r_square_score = []

for model in models:
    regressor = model().fit(x_train, y_train)
    pred = regressor.predict(x_test)
    mse.append(mean_squared_error(y_true= y_test, y_pred= pred))
    rmse.append(np.sqrt(mean_squared_error(y_true= y_test, y_pred= pred)))
    ex_var_score.append(explained_variance_score(y_true= y_test, y_pred= pred))
    r_square_score.append(r2_score(y_true= y_test, y_pred= pred))
```

```python
ML_model_df=pd.DataFrame({"Model":[ 'Linear Regression', 'Support Vector Regression', 'Decision Tree Regressor','Random Forest Regressor', 'Ridge', 'Lasso'],
                         "Mean Squarred Error":mse,
                         "Root Mean Squarred Error":rmse,
                         "Explained Variance Score":ex_var_score,
                         "R-Sqaure Score(Accuracy)":r_square_score})

ML_model_df.set_index('Model',inplace=True)
ML_model_df
```

### Conclusion

- Out of all the above models **Decision Tree Regressor** or **Random Forest Regressor** can be used for the predictive modelling. Even a little hyper-parameter tuning can be performed on them to improve the model performance
